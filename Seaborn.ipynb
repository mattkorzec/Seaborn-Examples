# Exploring Data with Seaborn

### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import os
os.chdir ('C:\\Users\\mkorzec\Downloads')
```

### Import and clean up data

```python
data = pd.read_csv('load_weather.csv')
cal = pd.read_csv('Calendar.csv')
```

```python
clean = pd.merge(data, cal, on = ['Year', 'Month', 'Day'])
```

```python
d1 = clean[clean['Holiday'] == 0]
```

```python
d1 = d1.drop(d1[d1.RainFall > 2].index)
```

### Distribution plots

```python
sns.distplot(d1['Load'])
```

```python
sns.jointplot(x = 'DewPoint', y = 'Temperature', data = d1, kind = 'hex')
```

```python
sns.pairplot(d1, vars = ['Load', 'Temperature', 'DewPoint', 'RainFall', 'WindSpeed'])
```

```python
sns.rugplot(d1['Load'])
```

### Categorical Plots

```python
sns.barplot(x = 'Hour', y = 'Load', data = d1)
```

```python
sns.barplot(x = 'Weekday', y = 'Load', data = d1, estimator = np.std)
```

```python
sns.countplot(x = 'Weekday', data = d1)
```

```python
sns.boxplot(x = 'Weekday', y = 'Load', data = d1)
```

```python
sns.violinplot(x = 'Weekday', y = 'Load', data = d1)
```

```python
sns.stripplot(x = 'Weekday', y = 'Load', data = d1, jitter = True)
```

```python
sns.swarmplot(x = 'Weekday', y = 'Load', data = d1)
```

### Matrix Plots

```python
tc = d1.corr()
```

```python
sns.heatmap(tc, annot = True, cmap = 'magma', linecolor = 'white', linewidths = 2)
```

### Grids

```python
g = sns.PairGrid(d1, vars = ['Load', 'Temperature', 'DewPoint', 'RainFall', 'WindSpeed'])
```

### Regression plots

```python
sns.lmplot(x = 'Temperature', y = 'Load', data = d1, markers = 'o')
```

```python
sns.lmplot(x = 'Temperature', y = 'Load', data = d1, col = 'Weekday')
```

### Style and Color

```python
sns.set_style('whitegrid')
sns.countplot(x = 'Weekday', data = d1)
sns.despine()
```

```python
plt.figure(figsize = (12,3))
sns.countplot(x = 'Weekday', data = d1)
```

```python
sns.set_context('poster', font_scale = 1.5)
sns.countplot(x = 'Weekday', data = d1, palette = 'cool')
```
